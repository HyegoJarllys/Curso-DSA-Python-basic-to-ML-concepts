#### Exercício 1

Escreva uma função que receba os valores de três lados de um triângulo e o classifique como "Equilátero" (todos os lados iguais), "Isósceles" (dois lados iguais) ou "Escaleno" (todos os lados diferentes).

```python
def triangulo (lado1, lado2, lado3):
    if (lado1 + lado2 <= lado3) or (lado1 + lado3 <= lado2) or (lado2 + lado3 <= lado1):
        return "Os lados não formam um triangulo"
    elif lado1 == lado2 == lado3:
        return "O triangulo é Equilátero"
    elif lado1 != lado2 and lado2 != lado3 and lado1 != lado3:
        return "O triangulo é Escaleno"
    else:
        return "O triangulo é Isósceles"
    
lado1 = float(input("tamanho do lado 1?"))
lado2 = float(input("tamanho do lado 2?"))
lado3 = float(input("tamanho do lado 3?"))
print(triangulo (lado1, lado2, lado3))
```

#### Exercício 2

Escreva uma função que recebe um número inteiro e exibe a tabuada de multiplicação desse número, do 1 ao 10.

```python
def tabuada(numero):
    for i in range(1, 11):
        print(f"{numero} x {i} = {numero * i}")

numero = int(input("Digite um número para ver a tabuada: "))
print(f"Tabuada do {numero} é :")
tabuada(numero)
```

#### Exercício 3 

Você recebeu um dicionário com os nomes dos alunos e suas respectivas notas. Escreva uma função que calcula a média da turma e retorna uma lista com os nomes dos alunos que tiveram nota acima da média.

```python
alunos = {"José": 8.5, "Maria": 7.2, "João": 9.0, "Ana": 6.8}

def nota_final (alunos):
    
    media = sum(alunos.values()) / len(alunos)
    
    alunos_acima_media = [nome for nome, nota in alunos.items() if nota > media]
    return print(f"Os alunos com nota acima da média são: {alunos_acima_media}, a média da turma é: {media:.2f}")

nota_final(alunos)
```

#### Exercício 4

Dada uma lista de números, crie uma nova lista usando list comprehension que contenha o quadrado de cada número par da lista original.

```python

entrada = input("Digite números separados por espaço: ")
lista_numeros= list(map(int, entrada.split()))
nova_lista_numeros=[ x**2 for x in lista_numeros if x % 2 == 0]

print(nova_lista_numeros)
```

#### Exercício 5

Crie uma função chamada dsa_calcula_imc que aceite dois argumentos: peso (em kg) e altura (em metros). A função deve calcular o Índice de Massa Corporal (IMC) usando a fórmula IMC=peso/altura^2 e retornar o valor do IMC.

```python
def dsa_calcula_imc(peso, altura):
    imc = peso / (altura ** 2)
    print(f"Seu IMC é: {imc:.2f}")

peso = float(input("Digite seu peso em kg: EX: 70.5 "))
altura = float(input("Digite sua altura em metros: EX: 1.75 "))

dsa_calcula_imc(peso, altura)
```

#### Exercício 6

#### Exercício 7

Crie uma função que receba uma lista de números inteiros e retorne um dicionário contendo a contagem de quantos números são pares e quantos são ímpares.

```python
valores = input("Digite os valores da lista com espaço entre eles:")
lista_valores = list(map(int, valores.split()))


def receber (lista_valores):

    pares = [x for x in lista_valores if x % 2 == 0]
    npares = len(pares)
    impares = [x for x in lista_valores if x % 2 != 0]
    nimpares = len(impares)
    return { 'pares ': npares , 'impares': nimpares }
 

receber (lista_valores)
```

#### Exercício 8

Crie uma função que receba uma lista de strings (potenciais e-mails) e um parâmetro opcional dominio_desejado (com valor padrão "gmail.com"). A função deve retornar uma nova lista, usando list comprehension, contendo apenas os e-mails que terminam com o domínio desejado.

```python
lista_completa = input("Digite os emails separados por espaço:")
email = list(lista_completa.split())
def dsa_email(email,dominio_desejado="gmail.com"):
    
    emails_desejados = [x for x in email if x.endswith(f"@{dominio_desejado}")]
    return emails_desejados

dsa_email(email)
```
